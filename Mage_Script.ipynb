```python
from pyautogui import *
import pyautogui
import time
import keyboard
import random
import win32api, win32con

#Taskbar position-small = 105,1040,1260,40
#Taskbar position-large = 172,198,1250,140

#AutoOff-small = 1355,710,95,95
#ZeroPotion-small = 470,60,85,85
#ZeroScroll-small = 1500,865,115,110

#HP bar 80% position2-small = 380, 70
#HP bar 70% position-small = 350, 70
#HP bar 80% position-large = 530, 82
#HP bar 70% position-large = 500, 82
#Red RGB > 200

#Buff icon position-small = 147,158,900,75
#Buff icon position-large = 105,1040,1260,40

pix20 = pyautogui.pixel(215,70)
d_pad = ['a','w','s','d']
lt = time.localtime()

while 1:
    time.sleep(1)
    if pyautogui.locateOnScreen('BSIcon.png', region=(105,1040,1260,40)) != None:
        print('BS on top')
        if pyautogui.locateOnScreen('Town.png', region=(1630,430,120,70), grayscale=True, confidence=0.8) == None:
            print('Not in town')
            if pyautogui.locateOnScreen('ZeroPotion.png', region=(470,60,85,85)) != None:
                print('No potion, return to town.')
                pyautogui.hotkey('7')
                time.sleep(5)
            elif pyautogui.locateOnScreen('ZeroScroll.png', region=(1500,865,115,110)) != None:
                print('No scroll, return to town.')
                pyautogui.hotkey('7')
                time.sleep(5)
            elif pix20[0] < 20:
                print('HP: ', pix20[0])
                pyautogui.hotkey('7')
                time.sleep(5)
            elif (lt.tm_hour == 9 or lt.tm_hour == 12 or lt.tm_hour == 18 or lt.tm_hour == 21) \
                and lt.tm_min == 1:
                print('Check mail')
                pyautogui.hotkey('shift','m')
                time.sleep(5)
            elif pyautogui.locateOnScreen('WorldBoss.png', region=(1000,30,350,140), confidence=0.7) != None:
                print('World boss')
                pyautogui.hotkey('7')
                pyautogui.hotkey('shift','b')
                time.sleep(1200)
                pyautogui.hotkey('7')
        else:
            print('In town')
            pyautogui.hotkey('shift','h')
            time.sleep(10)
            step = 1
            while step == 1:
                if pyautogui.locateOnScreen('JunkDealer.png', region=(210,355,1210,345), grayscale=True, confidence=0.6) == None:
                    print('Look for junk dealer')
                    move = random.sample(d_pad,1)
                    print('w')
                    pyautogui.keyDown('w')
                    time.sleep(0.3)
                    pyautogui,keyUp('w')
                else:
                    print('Found junk dealer')
                    x, y = pyautogui.locateCenterOnScreen('JunkDealer.png', region=(210,355,1210,345), grayscale=True, confidence=0.6)
                    pyautogui.click(x, y-120)
                    time.sleep(2)
                    pyautogui.hotkey('shift','p')
                    time.sleep(5)
                    step = 2
            while step == 2:
                if pyautogui.locateOnScreen('Warehouse.png', region=(210,355,1210,345), grayscale=True, confidence=0.6) == None:
                    print('Look for warehouse')
                    move = random.sample(d_pad,1)
                    print('w')
                    pyautogui.keyDown('w')
                    time.sleep(0.3)
                    pyautogui,keyUp('w')
                else:
                    print('Found warehouse')
                    x, y = pyautogui.locateCenterOnScreen('Warehouse.png', region=(210,355,1210,345), grayscale=True, confidence=0.6)
                    pyautogui.click(x, y-120)
                    time.sleep(2)
                    pyautogui.hotkey('shift','p')
                    time.sleep(5)
                    step = 3
            while step == 3:
                if pyautogui.locateOnScreen('Grocery.png', region=(210,355,1210,345), grayscale=True, confidence=0.6) == None:
                    print('Look for grocery')
                    move = random.sample(d_pad,1)
                    print('w')
                    pyautogui.keyDown('w')
                    time.sleep(0.3)
                    pyautogui,keyUp('w')
                else:
                    print('Found grocery')
                    x, y = pyautogui.locateCenterOnScreen('Grocery.png', region=(210,355,1210,345), grayscale=True, confidence=0.6)
                    pyautogui.click(x, y-120)
                    time.sleep(2)
                    pyautogui.hotkey('shift','p')
                    time.sleep(5)
                    pyautogui.hotkey('shift','f')
                    step = 0
                    time.sleep(5)
```
